```python
# =========================================================
# Multi-seed experiment (10 seeds, repeated per topology)
# Report only the FINAL repeat per seed so results are comparable
# to the older one-point-per-seed experiments.
# =========================================================

import matplotlib.pyplot as plt
import pandas as pd

from opinion_dynamics.experiments.online_repeated import (
    run_multi_seed_repeated_experiment_dynamics,
    run_repeated_online_id_on_env,
)

seeds = range(10)
B_campaign = 1.0
num_campaigns_total = 5
FIT_MAX_STEPS = 5_000
NUM_REPEATS = 100

# Laplacian
_df_lap_all = run_multi_seed_repeated_experiment_dynamics(
    seeds=seeds,
    repeats_per_seed=NUM_REPEATS,
    dynamics_model="laplacian",
    B_campaign=B_campaign,
    num_campaigns_total=num_campaigns_total,
    suppress_fit_logs=True,
    fit_max_steps=FIT_MAX_STEPS,
    fit_mae_stop=1e-2,
    fit_batch_size=256,
    fit_check_every=200,
)

# COCA
_df_coca_all = run_multi_seed_repeated_experiment_dynamics(
    seeds=seeds,
    repeats_per_seed=NUM_REPEATS,
    dynamics_model="coca",
    B_campaign=B_campaign,
    num_campaigns_total=num_campaigns_total,
    suppress_fit_logs=True,
    fit_max_steps=FIT_MAX_STEPS,
    fit_mae_stop=2e-2,
    fit_batch_size=512,
    fit_check_every=200,
)

# Keep only the final accumulated repeat per seed.
df_lap = (
    _df_lap_all.sort_values(["seed", "repeat"]) 
    .groupby("seed", as_index=False, group_keys=False)
    .tail(1)
    .reset_index(drop=True)
)
df_coca = (
    _df_coca_all.sort_values(["seed", "repeat"]) 
    .groupby("seed", as_index=False, group_keys=False)
    .tail(1)
    .reset_index(drop=True)
)

df = pd.concat([df_lap, df_coca], axis=0).reset_index(drop=True)
display(df)

print("=== Aggregate by dynamics (final repeat only) ===")
cols = ["repeat", "train_pairs_total", "v_L1_final","A_MAE_final","mean_gap_to_oracle_end","vx_gap_to_oracle_end","time_fit_inner"]
display(df.groupby("dynamics")[cols].describe().T)

plt.figure(figsize=(9,3))
for dyn in ["laplacian", "coca"]:
    sub = df[df["dynamics"] == dyn].sort_values("seed")
    plt.plot(sub["seed"], sub["A_MAE_final"], marker="o", label=f"{dyn}: A_MAE")
plt.xlabel("seed"); plt.ylabel("MAE(A_hat, A_true)")
plt.title("Adjacency MAE vs seed (final accumulated repeat)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

plt.figure(figsize=(9,3))
for dyn in ["laplacian", "coca"]:
    sub = df[df["dynamics"] == dyn].sort_values("seed")
    plt.plot(sub["seed"], sub["v_L1_final"], marker="o", label=f"{dyn}: v_L1")
plt.xlabel("seed"); plt.ylabel("||v_hat - v_true||_1")
plt.title("Centrality error vs seed (final accumulated repeat)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from opinion_dynamics.experiments.plots import (
    plot_impulse_node_trajectories,
    show_matrix_with_cell_grid,
    concat_intermediate,
)
from opinion_dynamics.experiments.metrics import graph_sanity
from opinion_dynamics.experiments.rollouts import (
    rollout_with_v_intermediate,
    make_env_with_dynamics,
)
from opinion_dynamics.experiments.online_repeated import run_repeated_online_id_on_env
from opinion_dynamics.utils.env_setup import EnvironmentFactory


# =========================================================
# Settings
# =========================================================
SEED_TO_PLOT = 10
DYNAMICS = "coca"   # or "laplacian"
NUM_REPEATS = 100
REPEAT_TO_PLOT = NUM_REPEATS - 1   # final accumulated repeat

FIT_KW = dict(
    fit_max_steps=FIT_MAX_STEPS,
    fit_mae_stop=1e-2 if DYNAMICS == "coca" else 1e-3,
    fit_batch_size=512 if DYNAMICS == "coca" else 64,
    fit_check_every=200,
)

env_factory = EnvironmentFactory()


def make_eval_env():
    return make_env_with_dynamics(
        env_factory,
        seed=SEED_TO_PLOT,
        dynamics_model=DYNAMICS,
    )


# =========================================================
# 1) Train with accumulated experience across many repeats
#    on the same graph topology
# =========================================================
env_template = make_eval_env()

out = run_repeated_online_id_on_env(
    env_template,
    num_repeats=NUM_REPEATS,
    B_campaign=B_campaign,
    num_campaigns_total=num_campaigns_total,
    lr=1e-3,
    l2_lambda=0.0,
    device="cpu",
    update_A_each_campaign=True,
    suppress_fit_logs=True,
    **FIT_KW,
)

metrics = out["rows"][REPEAT_TO_PLOT]
art = out["artifacts"][REPEAT_TO_PLOT]

print(f"=== REPEAT {REPEAT_TO_PLOT + 1}/{NUM_REPEATS} ===")
print(f"train_pairs_total: {metrics['train_pairs_total']}")
print("=== METRICS ===")
for k, v in metrics.items():
    print(f"{k}: {v}")

env_learn = art["env"]

# =========================================================
# 2) Pull ground-truth + learned-final
# =========================================================
A_true = np.asarray(art["A_true"], dtype=float)
v_true = np.asarray(art["v_true"], dtype=float)

A_hat_final = np.asarray(art["A_hat_final"], dtype=float)
v_hat_final = np.asarray(art["v_hat_final"], dtype=float)

# This is the actual trajectory from the final repeat.
# We keep it only for reference if you want to inspect it later.
states_learn_actual = np.asarray(art["states"], dtype=float)
N = states_learn_actual.shape[1]
K_total = states_learn_actual.shape[0] - 1

# =========================================================
# 3) Build a reproducible evaluation x0 from the ORIGINAL seed
#    so plots are directly comparable to the old single-run experiment
# =========================================================
env_x0 = make_eval_env()
x0_plot, _ = env_x0.reset()
x0_plot = np.asarray(x0_plot, dtype=float)
dt = float(getattr(env_x0, "t_s", 1.0))

# =========================================================
# 4) GRAPH SANITY (true vs learned-final)
# =========================================================
print("\n=== GRAPH SANITY (TRUE) ===")
for k, v in graph_sanity(A_true).items():
    print(f"{k}: {v}")

print("\n=== GRAPH SANITY (LEARNED FINAL) ===")
for k, v in graph_sanity(A_hat_final).items():
    print(f"{k}: {v}")

# =========================================================
# 5) Heatmaps
# =========================================================
show_matrix_with_cell_grid(
    A_true,
    title=f"True A ({DYNAMICS}, seed={SEED_TO_PLOT})",
    grid_alpha=0.25,
    grid_lw=0.6,
)
show_matrix_with_cell_grid(
    A_hat_final,
    title=f"Learned A_hat final ({DYNAMICS}, seed={SEED_TO_PLOT}, final repeat)",
    grid_alpha=0.25,
    grid_lw=0.6,
)

# =========================================================
# 6) Evaluate all comparison curves on the SAME common x0
#    using a FRESH env for each rollout
# =========================================================
le_out = rollout_with_v_intermediate(
    make_eval_env(),
    x0_plot,
    K_total,
    B_campaign,
    v_hat_final,
    zero_first_campaign=True,
)

or_out = rollout_with_v_intermediate(
    make_eval_env(),
    x0_plot,
    K_total,
    B_campaign,
    v_true,
    zero_first_campaign=True,
)

nc_out = rollout_with_v_intermediate(
    make_eval_env(),
    x0_plot,
    K_total,
    B_campaign,
    None,
    zero_first_campaign=True,
)

# =========================================================
# 7) "Wrong places" baseline: push budget toward least-central nodes
# =========================================================
order = np.argsort(v_true)  # ascending: least central first
ranks = np.empty_like(order, dtype=float)
ranks[order] = np.arange(1, len(v_true) + 1)  # 1..N
v_anti = 1.0 / ranks
v_anti = v_anti / v_anti.sum()

anti_out = rollout_with_v_intermediate(
    make_eval_env(),
    x0_plot,
    K_total,
    B_campaign,
    v_anti,
    zero_first_campaign=True,
)

# =========================================================
# 8) Fine-grained trajectories
# =========================================================
X_le, T_le = concat_intermediate(
    le_out["intermediate_states_list"],
    le_out["intermediate_times_list"],
    dt=dt,
)
X_or, T_or = concat_intermediate(
    or_out["intermediate_states_list"],
    or_out["intermediate_times_list"],
    dt=dt,
)

plot_impulse_node_trajectories(
    X_or,
    T_or,
    title=f"Node trajectories (fine-grained) — ORACLE true v ({DYNAMICS}, seed={SEED_TO_PLOT})",
)
plot_impulse_node_trajectories(
    X_le,
    T_le,
    title=f"Node trajectories (fine-grained) — LEARNED final controller ({DYNAMICS}, seed={SEED_TO_PLOT})",
)

# =========================================================
# 9) Centrality comparison (NO sorting)
# =========================================================
print("\n=== CENTRALITY COMPARISON (by node index) ===")
print("L1(v_hat_final - v_true):", float(np.sum(np.abs(v_hat_final - v_true))))

idx = np.arange(N)
diff = v_hat_final - v_true

plt.figure(figsize=(10, 4))
plt.plot(idx, v_true, marker="o", label="v_true")
plt.plot(idx, v_hat_final, marker="o", label="v_hat_final")
plt.xlabel("node index i")
plt.ylabel("centrality value")
plt.title(f"Centrality comparison (true vs learned) — {DYNAMICS}, seed={SEED_TO_PLOT}")
plt.xticks(idx)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

plt.figure(figsize=(10, 3))
plt.axhline(0.0, linewidth=1)
plt.plot(idx, diff, marker="o")
plt.xlabel("node index i")
plt.ylabel("v_hat - v_true")
plt.title(f"Centrality error per node — {DYNAMICS}, seed={SEED_TO_PLOT}")
plt.xticks(idx)
plt.grid(True, alpha=0.3)
plt.show()

# =========================================================
# 10) Mean opinion plot at campaign boundaries
# =========================================================
states_learn = np.asarray(le_out["states"], dtype=float)
states_oracle = np.asarray(or_out["states"], dtype=float)
states_noc = np.asarray(nc_out["states"], dtype=float)
states_anti = np.asarray(anti_out["states"], dtype=float)

mean_learn = states_learn.mean(axis=1)
mean_or = states_oracle.mean(axis=1)
mean_nc = states_noc.mean(axis=1)
mean_anti = states_anti.mean(axis=1)

plt.figure(figsize=(8, 3))
plt.plot(mean_learn, marker="o", label="learned final controller")
plt.plot(mean_or, marker="o", label="oracle (true v)")
plt.plot(mean_anti, marker="o", label="wrong places (least-central)")
plt.plot(mean_nc, marker="o", label="no control")
plt.xlabel("Campaign boundary k")
plt.ylabel("mean opinion")
plt.title(f"Trajectory: mean opinion — {DYNAMICS}, seed={SEED_TO_PLOT} (common x0)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.ylim(0, 1)
plt.show()

print("\n=== CAMPAIGN-0 OVERLAP CHECK ===")
print(
    "learned states[1] vs oracle states[1] L_inf:",
    float(np.max(np.abs(states_learn[1] - states_oracle[1]))),
)
print(
    "learned states[1] vs nocontrol states[1] L_inf:",
    float(np.max(np.abs(states_learn[1] - states_noc[1]))),
)
```
