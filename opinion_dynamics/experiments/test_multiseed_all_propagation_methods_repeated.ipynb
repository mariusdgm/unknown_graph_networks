```python
# =========================================================
# Multi-seed experiment across ALL propagation methods
# (10 seeds, repeated per topology)
# Report only the FINAL repeat per seed so results are comparable
# to the older one-point-per-seed experiments.
# =========================================================

import matplotlib.pyplot as plt
import pandas as pd

from opinion_dynamics.experiments.online_repeated import (
    run_multi_seed_repeated_experiment_dynamics,
)

seeds = range(10)
B_campaign = 1.0
num_campaigns_total = 5
FIT_MAX_STEPS = 5_000
NUM_REPEATS = 100

DYNAMICS_LIST = [
    "laplacian",
    "coca",
    "friedkinjohnsen",
    "hegselmannkrause",
    "nonlinearinfluence",
    # "repulsion",
]

FIT_BY_DYNAMICS = {
    "laplacian": dict(fit_max_steps=FIT_MAX_STEPS, fit_mae_stop=1e-2, fit_batch_size=256, fit_check_every=200),
    "coca": dict(fit_max_steps=FIT_MAX_STEPS, fit_mae_stop=2e-2, fit_batch_size=512, fit_check_every=200),
    "friedkinjohnsen": dict(fit_max_steps=FIT_MAX_STEPS, fit_mae_stop=2e-2, fit_batch_size=256, fit_check_every=200),
    "hegselmannkrause": dict(fit_max_steps=FIT_MAX_STEPS, fit_mae_stop=2e-2, fit_batch_size=256, fit_check_every=200),
    "nonlinearinfluence": dict(fit_max_steps=FIT_MAX_STEPS, fit_mae_stop=2e-2, fit_batch_size=256, fit_check_every=200),
    # "repulsion": dict(fit_max_steps=FIT_MAX_STEPS, fit_mae_stop=3e-2, fit_batch_size=256, fit_check_every=200),
}

dfs_final = []
dfs_all = {}

for dyn in DYNAMICS_LIST:
    print(f"=== Running {dyn} ===")
    df_all = run_multi_seed_repeated_experiment_dynamics(
        seeds=seeds,
        repeats_per_seed=NUM_REPEATS,
        dynamics_model=dyn,
        B_campaign=B_campaign,
        num_campaigns_total=num_campaigns_total,
        suppress_fit_logs=True,
        **FIT_BY_DYNAMICS[dyn],
    )
    dfs_all[dyn] = df_all

    df_final = (
        df_all.sort_values(["seed", "repeat"])
        .groupby("seed", as_index=False, group_keys=False)
        .tail(1)
        .reset_index(drop=True)
    )
    dfs_final.append(df_final)

df = pd.concat(dfs_final, axis=0).reset_index(drop=True)
display(df)

print("=== Aggregate by dynamics (final repeat only) ===")
cols = [
    "repeat",
    "train_pairs_total",
    "v_L1_final",
    "A_MAE_final",
    "mean_gap_to_oracle_end",
    "vx_gap_to_oracle_end",
    "time_fit_inner",
]
display(df.groupby("dynamics")[cols].describe().T)

for metric, ylabel, title in [
    ("A_MAE_final", "MAE(A_hat, A_true)", "Adjacency MAE vs seed (final accumulated repeat)"),
    ("v_L1_final", "||v_hat - v_true||_1", "Centrality error vs seed (final accumulated repeat)"),
    ("mean_gap_to_oracle_end", "oracle gap", "Mean gap to oracle vs seed (final accumulated repeat)"),
]:
    plt.figure(figsize=(10, 4))
    for dyn in DYNAMICS_LIST:
        sub = df[df["dynamics"] == dyn].sort_values("seed")
        plt.plot(sub["seed"], sub[metric], marker="o", label=dyn)
    plt.xlabel("seed")
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(True, alpha=0.3)
    plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from opinion_dynamics.experiments.plots import (
    plot_impulse_node_trajectories,
    show_matrix_with_cell_grid,
    concat_intermediate,
)
from opinion_dynamics.experiments.metrics import graph_sanity
from opinion_dynamics.experiments.rollouts import (
    rollout_with_v_intermediate,
    make_env_with_dynamics,
)
from opinion_dynamics.experiments.online_repeated import run_repeated_online_id_on_env
from opinion_dynamics.utils.env_setup import EnvironmentFactory


# =========================================================
# Settings (local names so notebook-wide config is not overwritten)
# =========================================================
SEED_TO_PLOT = 10
num_repeats_local = int(globals().get("NUM_REPEATS", 100))
repeat_to_plot = num_repeats_local - 1
num_campaigns_total_local = int(globals().get("num_campaigns_total", 5))
fit_max_steps_local = int(globals().get("FIT_MAX_STEPS", 5_000))

dynamics_list_local = list(globals().get("DYNAMICS_LIST", [
    "laplacian",
    "coca",
    "friedkinjohnsen",
    "hegselmannkrause",
    "nonlinearinfluence",
]))

fit_by_dynamics_local = globals().get("FIT_BY_DYNAMICS", {
    "laplacian": dict(fit_max_steps=fit_max_steps_local, fit_mae_stop=1e-2, fit_batch_size=256, fit_check_every=200),
    "coca": dict(fit_max_steps=fit_max_steps_local, fit_mae_stop=2e-2, fit_batch_size=512, fit_check_every=200),
    "friedkinjohnsen": dict(fit_max_steps=fit_max_steps_local, fit_mae_stop=2e-2, fit_batch_size=256, fit_check_every=200),
    "hegselmannkrause": dict(fit_max_steps=fit_max_steps_local, fit_mae_stop=2e-2, fit_batch_size=256, fit_check_every=200),
    "nonlinearinfluence": dict(fit_max_steps=fit_max_steps_local, fit_mae_stop=2e-2, fit_batch_size=256, fit_check_every=200),
})

env_factory = EnvironmentFactory()


def make_eval_env(dynamics_name):
    return make_env_with_dynamics(
        env_factory,
        seed=SEED_TO_PLOT,
        dynamics_model=dynamics_name,
    )

for dyn_name in dynamics_list_local:
    print("\n" + "=" * 80)
    print(f"=== DYNAMICS: {dyn_name} | SEED {SEED_TO_PLOT} ===")
    print("=" * 80)

    fit_kw = fit_by_dynamics_local[dyn_name]

    # =========================================================
    # 1) Train with accumulated experience across many repeats 
    #    on the same graph topology
    # =========================================================
    env_template = make_eval_env(dyn_name)

    out = run_repeated_online_id_on_env(
        env_template,
        num_repeats=num_repeats_local,
        B_campaign=B_campaign,
        num_campaigns_total=num_campaigns_total_local,
        lr=1e-3,
        l2_lambda=0.0,
        device="cpu",
        update_A_each_campaign=False,
        suppress_fit_logs=True,
        **fit_kw,
    )

    metrics = out["rows"][repeat_to_plot]
    art = out["artifacts"][repeat_to_plot]

    print(f"=== REPEAT {repeat_to_plot + 1}/{num_repeats_local} ===")
    print(f"train_pairs_total: {metrics['train_pairs_total']}")
    print("=== METRICS ===")
    for k, v in metrics.items():
        print(f"{k}: {v}")

    # =========================================================
    # 2) Pull ground-truth + learned-final
    # =========================================================
    A_true = np.asarray(art["A_true"], dtype=float)
    v_true = np.asarray(art["v_true"], dtype=float)

    A_hat_final = np.asarray(art["A_hat_final"], dtype=float)
    v_hat_final = np.asarray(art["v_hat_final"], dtype=float)

    states_learn_actual = np.asarray(art["states"], dtype=float)
    N = states_learn_actual.shape[1]
    K_total = states_learn_actual.shape[0] - 1

    # =========================================================
    # 3) Build a reproducible evaluation x0 from the ORIGINAL seed
    # =========================================================
    env_x0 = make_eval_env(dyn_name)
    x0_plot, _ = env_x0.reset()
    x0_plot = np.asarray(x0_plot, dtype=float)
    dt = float(getattr(env_x0, "t_s", 1.0))

    # =========================================================
    # 4) GRAPH SANITY (true vs learned-final)
    # =========================================================
    print("\n=== GRAPH SANITY (TRUE) ===")
    for k, v in graph_sanity(A_true).items():
        print(f"{k}: {v}")

    print("\n=== GRAPH SANITY (LEARNED FINAL) ===")
    for k, v in graph_sanity(A_hat_final).items():
        print(f"{k}: {v}")

    show_matrix_with_cell_grid(
        A_true,
        title=f"True A ({dyn_name}, seed={SEED_TO_PLOT})",
        grid_alpha=0.25,
        grid_lw=0.6,
    )
    show_matrix_with_cell_grid(
        A_hat_final,
        title=f"Learned A_hat final ({dyn_name}, seed={SEED_TO_PLOT}, final repeat)",
        grid_alpha=0.25,
        grid_lw=0.6,
    )

    # =========================================================
    # 5) Evaluate all comparison curves on the SAME common x0
    # =========================================================
    le_out = rollout_with_v_intermediate(
        make_eval_env(dyn_name),
        x0_plot,
        K_total,
        B_campaign,
        v_hat_final,
        zero_first_campaign=True,
    )

    or_out = rollout_with_v_intermediate(
        make_eval_env(dyn_name),
        x0_plot,
        K_total,
        B_campaign,
        v_true,
        zero_first_campaign=True,
    )

    nc_out = rollout_with_v_intermediate(
        make_eval_env(dyn_name),
        x0_plot,
        K_total,
        B_campaign,
        None,
        zero_first_campaign=True,
    )

    order = np.argsort(v_true)
    ranks = np.empty_like(order, dtype=float)
    ranks[order] = np.arange(1, len(v_true) + 1)
    v_anti = 1.0 / ranks
    v_anti = v_anti / v_anti.sum()

    anti_out = rollout_with_v_intermediate(
        make_eval_env(dyn_name),
        x0_plot,
        K_total,
        B_campaign,
        v_anti,
        zero_first_campaign=True,
    )

    # =========================================================
    # 6) Fine-grained trajectories
    # =========================================================
    X_le, T_le = concat_intermediate(
        le_out["intermediate_states_list"],
        le_out["intermediate_times_list"],
        dt=dt,
    )
    X_or, T_or = concat_intermediate(
        or_out["intermediate_states_list"],
        or_out["intermediate_times_list"],
        dt=dt,
    )

    plot_impulse_node_trajectories(
        X_or,
        T_or,
        title=f"Node trajectories (fine-grained) — ORACLE true v ({dyn_name}, seed={SEED_TO_PLOT})",
    )
    plot_impulse_node_trajectories(
        X_le,
        T_le,
        title=f"Node trajectories (fine-grained) — LEARNED final controller ({dyn_name}, seed={SEED_TO_PLOT})",
    )

    # =========================================================
    # 7) Centrality comparison (NO sorting)
    # =========================================================
    print("\n=== CENTRALITY COMPARISON (by node index) ===")
    print("L1(v_hat_final - v_true):", float(np.sum(np.abs(v_hat_final - v_true))))

    idx = np.arange(N)
    diff = v_hat_final - v_true

    plt.figure(figsize=(10, 4))
    plt.plot(idx, v_true, marker="o", label="v_true")
    plt.plot(idx, v_hat_final, marker="o", label="v_hat_final")
    plt.xlabel("node index i")
    plt.ylabel("centrality value")
    plt.title(f"Centrality comparison (true vs learned) — {dyn_name}, seed={SEED_TO_PLOT}")
    plt.xticks(idx)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()

    plt.figure(figsize=(10, 3))
    plt.axhline(0.0, linewidth=1)
    plt.plot(idx, diff, marker="o")
    plt.xlabel("node index i")
    plt.ylabel("v_hat - v_true")
    plt.title(f"Centrality error per node — {dyn_name}, seed={SEED_TO_PLOT}")
    plt.xticks(idx)
    plt.grid(True, alpha=0.3)
    plt.show()

    # =========================================================
    # 8) Mean opinion plot at campaign boundaries
    # =========================================================
    states_learn = np.asarray(le_out["states"], dtype=float)
    states_oracle = np.asarray(or_out["states"], dtype=float)
    states_noc = np.asarray(nc_out["states"], dtype=float)
    states_anti = np.asarray(anti_out["states"], dtype=float)

    mean_learn = states_learn.mean(axis=1)
    mean_or = states_oracle.mean(axis=1)
    mean_nc = states_noc.mean(axis=1)
    mean_anti = states_anti.mean(axis=1)

    plt.figure(figsize=(8, 3))
    plt.plot(mean_learn, marker="o", label="learned final controller")
    plt.plot(mean_or, marker="o", label="oracle (true v)")
    plt.plot(mean_anti, marker="o", label="wrong places (least-central)")
    plt.plot(mean_nc, marker="o", label="no control")
    plt.xlabel("Campaign boundary k")
    plt.ylabel("mean opinion")
    plt.title(f"Trajectory: mean opinion — {dyn_name}, seed={SEED_TO_PLOT} (common x0)")
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.ylim(0, 1)
    plt.show()

    print("\n=== CAMPAIGN-0 OVERLAP CHECK ===")
    print(
        "learned states[1] vs oracle states[1] L_inf:",
        float(np.max(np.abs(states_learn[1] - states_oracle[1]))),
    )
    print(
        "learned states[1] vs nocontrol states[1] L_inf:",
        float(np.max(np.abs(states_learn[1] - states_noc[1]))),
    )
```
